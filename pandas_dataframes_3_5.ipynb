```
import pandas as pd
from scipy import stats
import seaborn as sns
import warnings
import matplotlib.pyplot as plt

sns.set()
warnings.filterwarnings('ignore')

data_file = '/content/drive/My Drive/Thinkful Data Science/python_for_data_scientists/data/houseprices.csv'

houseprices = pd.read_csv(data_file)
saleprice_ax = sns.distplot(houseprices['saleprice'], bins=10, kde=False)
for label in saleprice_ax.get_xticklabels():
    label.set_rotation(60)
```

```
fig = plt.figure()
fig.set_size_inches(8, 6)
lotarea_ax = sns.distplot(houseprices['lotarea'], bins=20)
```

```
fig = plt.figure()
fig.set_size_inches(10, 8)
ax_box = sns.boxplot(y='saleprice', x='overallcond', data=houseprices)
```

```
ax_box = sns.boxplot(y='saleprice', x='overallcond',  hue='centralair', data=houseprices)
```

```
ax_scatter = sns.lmplot(x='lotarea', y='saleprice', data=houseprices)
stats.pearsonr(houseprices.lotarea, houseprices.saleprice)
```
