```python
bools = (kick_df['category'] == 'Documentary') & (kick_df['state'] == 'successful')
docs = kick_df.iloc[bools.values].sort_values(by='pledged', ascending=False)
docs.head(10)
```

```python
kick_df = kick_df.assign(average_per_backer = (kick_df.pledged / kick_df.backers))
kick_df[kick_df.backers == 0]
```

```python
kick_df = kick_df[kick_df.backers != 0]
kick_df = kick_df.assign(average_per_backer = (kick_df.pledged / kick_df.backers))
kick_df[kick_df.backers == 0]
```

```python
import pandas as pd

state_file = '/content/drive/My Drive/python_for_data_scientists/data/state-populations.csv'
div_file = '/content/drive/My Drive/python_for_data_scientists/data/census-divisions.csv'

state_pop = pd.read_csv(state_file)
census_div = pd.read_csv(div_file)

census = pd.merge(state_pop, census_div)
census.head(10)
```

```python
census_region = pd.melt(frame=census, id_vars=['state', 'region', 'division'],var_name='year',value_name='population')
census_region[census_region.year == '2012'].head(10)
```

```python
census_byyear = census_region.groupby(by=['year']).sum()
census_byyear.head(10)
```

```python
census_region = census_region.groupby(by=['region', 'division','year']).sum()
census_region.head(10)
```
