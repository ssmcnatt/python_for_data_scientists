```
# 6. Is there a statistically significant difference between the prices of
# houses that have an open porch versus houses that do not have an open porch? 
# How did you handle null values?

# First, let's see what data we have
prices_df[['openporchsf', 'saleprice']].info()
```

```
# There are no null values in the relevant columns.
# let us assume that a 0 in the openporchsf column means that there is no open porch

with_open_porches = prices_df[['openporchsf', 'saleprice']].iloc[(prices_df['openporchsf'] > 0).values]
with_open_porches.info()
```

```
without_open_porches = prices_df[['openporchsf', 'saleprice']].iloc[(prices_df['openporchsf'] == 0).values]
without_open_porches.info()
```

```
stats.ttest_ind(with_open_porches['saleprice'], without_open_porches['saleprice'])
```

Since the p value is very small we can conclude that this result is significant.
